OOPS

Inheritance is a mechanism in object-oriented programming (OOP) where a new class (derived class) is created from an existing class (base class). The derived class inherits attributes and methods from the base class, allowing for code reuse and establishing a hierarchical relationship between classes.

```python
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
    
    def display(self):
        print(f"The person is driving a car with {self.enginetype} engine")
```

```python
car1 = Car(4,2,"Petrol")
car1.display()
```

```python
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,is_selfdriving):
        super().__init__(windows,doors,enginetype)
        # super is used to inherit attributes of parent class
        self.is_selfdriving=is_selfdriving

    def selfdriving(self):
        print(f"Tesla supports self driving :  {self.is_selfdriving}")
```

```python
Tesla1 = Tesla(4,2,"electric",True)
Tesla1.selfdriving()
```

Multiple Inheritance

when a class inherits from more than one base class.

```python
class Father:
    def __init__(self):
        self.father_name = "John"
    
    def coding(self):
        print("Father knows coding")

class Mother:
    def __init__(self):
        self.mother_name = "Jane"
    
    def cooking(self):
        print("Mother knows cooking")

class Child(Father, Mother):  # inheriting from both Father and Mother
    def __init__(self):
        Father.__init__(self)
        Mother.__init__(self)
    
    def skills(self):
        print(f"Father: {self.father_name}, Mother: {self.mother_name}")
        self.coding()
        self.cooking()

c = Child()
c.skills()
```

Polymorphism means many forms

```python
#Method Overriding
class Animal:
    def sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

class Cow(Animal):
    def sound(self):
        print("Cow moos")

# Same method name, different behavior
animals = [Dog(), Cat(), Cow()]
for animal in animals:
    animal.sound()
```

```python
# Duck typing

class Car:
    def fuel(self):
        print("Car runs on petrol")

class Boat:
    def fuel(self):
        print("Boat runs on diesel")

class Plane:
    def fuel(self):
        print("Plane runs on jet fuel")

# No inheritance needed — just same method name
vehicles = [Car(), Boat(), Plane()]
for v in vehicles:
    v.fuel()
```

```python
# Method overloading

class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

calc = Calculator()
print(calc.add(2, 3))      
print(calc.add(2, 3, 4))   
```

```python
# Method with functions
class Shape:
    def area(self, a, b):
        print("Area of the shape")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width

c = Circle(7)
print(f"Area of Circle: {c.area()}")

r = Rectangle(10, 5)
print(f"Area of Rectangle: {r.area()}")
```

Encapsulation

Encapsulation means hiding the internal data of a class and restricting direct access to it. It bundles data and methods together and protects them from outside interference.

```python
#Public Members can be asseccible from anywhere

class Student:
    def __init__(self, name, age):
        self.name = name   # public
        self.age = age     # public

s = Student("Alice", 20)
print(s.name)  # Alice
print(s.age)   # 20
```

```python
#Protected Members Protected members are a convention — Python doesn't strictly enforce it, but it signals "don't access directly".

class Student:
    def __init__(self, name, age):
        self._name = name   # protected
        self._age = age     # protected

class GradStudent(Student):
    def show(self):
        print(f"Name: {self._name}, Age: {self._age}")  # accessible in subclass

g = GradStudent("Bob", 25)
g.show()           # Name: Bob, Age: 25
print(g._name)     # still accessible but not recommended
```

```python
#Private Members remains private 

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance   # private

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited {amount}. New balance: {self.__balance}")

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient funds")
        else:
            self.__balance -= amount
            print(f"Withdrawn {amount}. New balance: {self.__balance}")

    def get_balance(self):
        return self.__balance   # controlled access via method

acc = BankAccount("Alice", 1000)
acc.deposit(500)       
acc.withdraw(200)      
print(acc.get_balance())  
```

```python
# Getters and setters provide controlled access to private members

class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary

    # Getter
    def get_salary(self):
        return self.__salary

    # Setter
    def set_salary(self, salary):
        if salary < 0:
            print("Salary cannot be negative")
        else:
            self.__salary = salary

e = Employee("John", 50000)
print(e.get_salary())   
e.set_salary(60000)
print(e.get_salary())   
e.set_salary(-100)      
```

Abstraction

Abstraction means hiding the implementation details and only showing the essential features. You define a blueprint that other classes must follow.Python achieves abstraction using the ABC (Abstract Base Class) module.

```python
from abc import ABC, abstractmethod

class Animal(ABC):      # abstract class
    
    @abstractmethod
    def sound(self):    # abstract method — no implementation
        pass
    
    @abstractmethod
    def movement(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Dog barks")
    
    def movement(self):
        print("Dog runs on 4 legs")

class Bird(Animal):
    def sound(self):
        print("Bird chirps")
    
    def movement(self):
        print("Bird flies with wings")

d = Dog()
d.sound()       
d.movement()    

b = Bird()
b.sound()       
b.movement()    
```

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    
    @abstractmethod
    def authenticate(self):
        pass
    
    @abstractmethod
    def pay(self, amount):
        pass

class CreditCard(Payment):
    def authenticate(self):
        print("Authenticating Credit Card...")
    
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")

class UPI(Payment):
    def authenticate(self):
        print("Authenticating UPI PIN...")
    
    def pay(self, amount):
        print(f"Paid {amount} using UPI")

class Crypto(Payment):
    def authenticate(self):
        print("Authenticating Crypto Wallet...")
    
    def pay(self, amount):
        print(f"Paid {amount} using Cryptocurrency")

# Process payment without knowing internal details
def process_payment(payment, amount):
    payment.authenticate()
    payment.pay(amount)

process_payment(CreditCard(), 500)
process_payment(UPI(), 300)
process_payment(Crypto(), 1000)
```

Magic Methods

Magic methods (also called dunder methods) are special methods that start and end with double underscores __. Python calls them automatically in certain situations.

```python
class Student:
    def __init__(self, name, age):   # called when object is created
        self.name = name
        self.age = age
        print(f"Student {self.name} created")

    def __del__(self):               # called when object is deleted
        print(f"Student {self.name} deleted")

s = Student("Alice", 20)   # Student Alice created
del s                       # Student Alice deleted
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):      # for end users — print()
        return f"Student: {self.name}, Age: {self.age}"

    def __repr__(self):     # for developers — debugging
        return f"Student(name={self.name}, age={self.age})"

s = Student("Alice", 20)
print(str(s))    # Student: Alice, Age: 20
print(repr(s))   # Student(name=Alice, age=20)
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):       # +
        return Number(self.value + other.value)

    def __sub__(self, other):       # -
        return Number(self.value - other.value)

    def __mul__(self, other):       # *
        return Number(self.value * other.value)

    def __truediv__(self, other):   # /
        return Number(self.value / other.value)

    def __str__(self):
        return str(self.value)

a = Number(10)
b = Number(5)

print(a + b)   # 15
print(a - b)   # 5
print(a * b)   # 50
print(a / b)   # 2.0
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __eq__(self, other):    # ==
        return self.marks == other.marks

    def __lt__(self, other):    # 
        return self.marks < other.marks

    def __gt__(self, other):    # >
        return self.marks > other.marks

    def __le__(self, other):    # <=
        return self.marks <= other.marks

    def __ge__(self, other):    # >=
        return self.marks >= other.marks

    def __ne__(self, other):    # !=
        return self.marks != other.marks

s1 = Student("Alice", 90)
s2 = Student("Bob", 75)

print(s1 > s2)    # True
print(s1 < s2)    # False
print(s1 == s2)   # False
print(s1 != s2)   # True
```

```python
class Classroom:
    def __init__(self):
        self.students = []

    def add_student(self, name):
        self.students.append(name)

    def __len__(self):          # len()
        return len(self.students)

    def __contains__(self, name):   # 'in' operator
        return name in self.students

c = Classroom()
c.add_student("Alice")
c.add_student("Bob")
c.add_student("Charlie")

print(len(c))               # 3
print("Alice" in c)         # True
print("John" in c)          # False
```

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, number):     # allows object() call
        return number * self.factor

double = Multiplier(2)
triple = Multiplier(3)

print(double(5))    # 10
print(triple(5))    # 15
```

```python
class FileManager:
    def __init__(self, filename):
        self.filename = filename

    def __enter__(self):
        print(f"Opening {self.filename}")
        self.file = open(self.filename, 'w')
        return self.file

    def __exit__(self, exc_type, exc_val, exc_tb):
        print(f"Closing {self.filename}")
        self.file.close()

with FileManager("test.txt") as f:
    f.write("Hello World")
# Opening test.txt
# Closing test.txt
```

Operator overloading 

Operator overloading means giving extra meaning to existing operators for user-defined classes. The same operator behaves differently depending on the objects used.

```python
class Box:
    def __init__(self, length, width, height):
        self.length = length
        self.width = width
        self.height = height

    def volume(self):
        return self.length * self.width * self.height

    def __eq__(self, other):    # ==
        return self.volume() == other.volume()

    def __ne__(self, other):    # !=
        return self.volume() != other.volume()

    def __lt__(self, other):    # 
        return self.volume() < other.volume()

    def __gt__(self, other):    # >
        return self.volume() > other.volume()

    def __le__(self, other):    # <=
        return self.volume() <= other.volume()

    def __ge__(self, other):    # >=
        return self.volume() >= other.volume()

    def __str__(self):
        return f"Box(Volume={self.volume()})"

b1 = Box(2, 3, 4)    # volume = 24
b2 = Box(3, 3, 3)    # volume = 27

print(b1 == b2)   # False
print(b1 != b2)   # True
print(b1 < b2)    # True
print(b1 > b2)    # False
print(b1 <= b2)   # True
print(b1 >= b2)   # False
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):       # +
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):       # -
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):      # *
        return Vector(self.x * scalar, self.y * scalar)

    def __truediv__(self, scalar):  # /
        return Vector(self.x / scalar, self.y / scalar)

    def __floordiv__(self, scalar): # //
        return Vector(self.x // scalar, self.y // scalar)

    def __mod__(self, scalar):      # %
        return Vector(self.x % scalar, self.y % scalar)

    def __pow__(self, power):       # **
        return Vector(self.x ** power, self.y ** power)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(10, 20)
v2 = Vector(3, 5)

print(v1 + v2)      # Vector(13, 25)
print(v1 - v2)      # Vector(7, 15)
print(v1 * 2)       # Vector(20, 40)
print(v1 / 2)       # Vector(5.0, 10.0)
print(v1 // 3)      # Vector(3, 6)
print(v1 % 3)       # Vector(1, 2)
print(v2 ** 2)      # Vector(9, 25)
```

```python
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name}: ${self.price}"

class Cart:
    def __init__(self):
        self.items = []
        self.total = 0

    def __add__(self, item):        # cart + item
        self.items.append(item)
        self.total += item.price
        return self

    def __sub__(self, item):        # cart - item
        if item in self.items:
            self.items.remove(item)
            self.total -= item.price
        return self

    def __len__(self):              # len(cart)
        return len(self.items)

    def __str__(self):
        item_list = ", ".join([i.name for i in self.items])
        return f"Cart: [{item_list}] | Total: ${self.total}"

cart = Cart()
apple  = Item("Apple", 2)
bread  = Item("Bread", 3)
milk   = Item("Milk", 5)

cart = cart + apple
cart = cart + bread
cart = cart + milk
print(cart)         # Cart: [Apple, Bread, Milk] | Total: $10
print(len(cart))    # 3

cart = cart - bread
print(cart)         # Cart: [Apple, Milk] | Total: $7
print(len(cart))    # 2
```
