```python
number = int(input("Enter a number: "))
if number % 2 == 0:
    print(f"{number} is even.")
else:
    print(f"{number} is odd.")
```

```python
year = int(input("Enter a year: "))
if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
def print_items(*args, **kwargs):
    for item in args:
        print(f"positional arg: {item}")

    for key, value in kwargs.items():
        print(f"keyword args {key}: {value}")
```

```python
print_items(1, 2, 3, "Jacob" , 3.14566, name="Alice", age=30, city="New York")  
```

```python
def strong_password(password):
    if len(password) < 8:
        return False
    for char in password:
        if char.isdigit():
            return True
        if char.isupper():
            return True
        if char.islower():
            return True
        if not char.isalnum():
            return True
        if char.isspace():
            return True
    return False

print(strong_password("Pass123"))  # Expected: True
print(strong_password("Wejbbjfbj)(958)"))          # Expected: False
```
